```python
import pandas as pd

# Reading the csv file
Student = pd.read_csv("/content/StudentsPerformance.csv")
print(Student)
```

```python
# Getting the info of complete csv
print(Student.describe())
```

```python
# Creating a data frame
Student_Data = pd.DataFrame(Student)
print(Student_Data)
```

```python
# Understanding the data and dimensions
print(Student_Data.columns)
print(Student_Data.shape)
```

```python

# Getting the top rows
print(Student_Data.head())

```

```python
# Getting the bottom rows
print(Student_Data.tail())

```

```python
# Getting the Data Structure
print(Student_Data.info())
```

```python
# Converting vector object to factor object-with levels of factor
Student_Data['gender'] = pd.Categorical(Student_Data['gender'])
Student_Data['race/ethnicity'] = pd.Categorical(Student_Data['race/ethnicity'])
Student_Data['parental level of education'] = pd.Categorical(Student_Data['parental level of education'])
Student_Data['lunch'] = pd.Categorical(Student_Data['lunch'])
Student_Data['test preparation course'] = pd.Categorical(Student_Data['test preparation course'])

print(Student_Data.info())
```

```python
# Removing the unnecessary columns from the data
del Student_Data['lunch']
print(Student_Data.describe())

# Finding out the missing values if any
print(Student_Data.isna().sum())

# Save the clean Data
Student_Data.to_csv("clean_StudentsData.csv", index=False)
```

```python
import dataclasses
import pandas as pd
import seaborn as sns

data = pd.read_csv("/content/clean_StudentsData.csv")

# Convert the data to a Pandas DataFrame
df = pd.DataFrame(data)

# Create a pairplot using seaborn
sns.pairplot(df, hue='gender')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Create a scatterplot
sns.scatterplot(data=data, x="math score", y="reading score", hue="gender")
plt.title("Math Score vs Reading Score by Gender")
plt.xlabel("Math Score")
plt.ylabel("Reading Score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a boxplot
sns.boxplot(data=data, x="race/ethnicity", y="writing score", hue="race/ethnicity")
plt.title("Writing Score by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Writing Score")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a barplot
sns.countplot(data=data, x="test preparation course", hue="gender")
plt.title("Test Preparation Completion by Gender")
plt.xlabel("Test Preparation Course")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
student_data = data

# Create a boxplot
sns.boxplot(data=student_data, x="race/ethnicity", y="math score", hue="race/ethnicity")
plt.title("Math Score by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")
plt.show()

sns.boxplot(data=student_data, x="race/ethnicity", y="reading score", hue="race/ethnicity")
plt.title("Reading Score by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Reading Score")
plt.show()

sns.boxplot(data=student_data, x="race/ethnicity", y="writing score", hue="race/ethnicity")
plt.title("Writing Score by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Writing Score")
plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

# read in the data
Student =data

# create a table of test preparation status by gender
prep_table = pd.crosstab(Student['gender'], Student['test preparation course'])

# plot a pie chart for female students
female_prep = prep_table.loc['female']
plt.pie(female_prep, labels=female_prep.index, autopct='%1.1f%%')
plt.title("Test Preparation Status for Female Students")
plt.show()
```
