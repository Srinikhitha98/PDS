```python
import pandas as pd

# Read the raw data file
df = pd.read_csv('/content/rawdata.csv')

# Clean up the data
df = df.dropna() # Drop any rows with missing values
df = df.reset_index(drop=True) # Reset the row index
df = df.rename(columns={'Height (Inches)': 'Height', 'Weight (Pounds)': 'Weight'})#Rename the names 
df['Frailty'] = df['Frailty'].replace({'Y': True, 'N': False}) # Convert the 'Frailty' column to boolean values


# Save the cleaned data to a new file
df.to_csv('/content/cleaned_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
# Read the CSV file into a DataFrame
df = pd.read_csv('/content/rawdata.csv')

# Replace the column names
df = df.rename(columns={'Height (Inches)': 'Height', 'Weight (Pounds)': 'Weight'})

# Convert the "Frailty" column to a binary integer column
df['Frailty'] = (df['Frailty'] == 'Y').astype(int)

# Normalize the "Height" and "Weight" columns
df['Height'] = (df['Height'] - df['Height'].mean()) / df['Height'].std()
df['Weight'] = (df['Weight'] - df['Weight'].mean()) / df['Weight'].std()

# Create a pairplot of the data
sns.pairplot(df, hue='Frailty')

# Show the plot
plt.show()
```

```python
sns.boxplot(x='Frailty', y='Grip strength', data=df)
plt.xlabel('Frailty')
plt.ylabel('Grip strength')
plt.title('Boxplot of Grip Strength by Frailty')
plt.show()
```

```python
sns.scatterplot(x='Weight', y='Height', hue='Frailty', data=df)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Scatterplot of Weight vs. Height')
plt.show()
```
